```python
# k = 5 
#to find the k value 
#ELBOW METHOD 

from sklearn.cluster import KMeans 
sse = []  #sum of squared error
k = range(2,15) #take k values in between 2 to 15 
for i in k:
  model_demo =KMeans(n_clusters=i,random_state = 0)
  model_demo.fit(x)
  sse.append(model_demo.inertia_) #takes the errors and appends it to sse list which we have created 
plt.scatter(k,sse) #inertia_ is the error value
plt.plot(k,sse)
plt.show() 
#elbow is most prominent at 5 wrt x axis 
# hence we take k = 5
```

```python
k = 5
from sklearn.cluster import KMeans
model = KMeans(n_clusters=k,random_state=0)
model.fit(x)
y = model.predict(x)
y
```

```python
np.unique(y,return_counts=True)
#this tell us that there are 5 clusters and these are the no. of values for each cluster
```

```python
#plot the graph including the clusters
plt.figure(figsize = (15,5))
for i in range(k):
  plt.scatter(x[y == i,0], x[y == i,1]) #this line assigns separate colors to each cluster 
```

```python
#plot the graph including the clusters
plt.figure(figsize = (15,5))
for i in range(k):
  plt.scatter(x[y == i,0], x[y == i,1]) #this line assigns separate colors to each cluster 
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],s= 200) #creates the centroid by using the trained model 
plt.show()
```
