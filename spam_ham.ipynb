```python
import pandas as pd
df = pd.read_table('https://raw.githubusercontent.com/diazonic/Machine-Learning-using-sklearn/master/Datasets/spam.tsv')
df
```

```python
df.info()
```

```python
df
```

```python
df['label'].value_counts()
```

```python
x = df['message'].values
y = df['label'].values
```

```python
x
```

```python
y
```

```python
#train test split 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train, y_test = train_test_split(x,y,random_state = 0)
```

```python
x.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
vect = CountVectorizer(stop_words = 'english')
x_train_vect = vect.fit_transform(x_train)
x_test_vect = vect.transform(x_test)
```

```python
from sklearn.svm import SVC #support vector machine 
#SVC - support vector classifier 
#svc - similar to logistic regression 
#svc gives better accuracy 
```

```python
model = SVC()
model.fit(x_train_vect, y_train) 
```

```python
y_pred = model.predict(x_test_vect)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```

```python
#USING A PIPELINE (pipeline is used to combine two estimators / two different processes together) 
# pipeline includes fit, predict, metrics
```

```python
#PIPELINE - SVC + count vectorizer 
from sklearn.pipeline import make_pipeline
text_model = make_pipeline(CountVectorizer(),SVC())

text_model.fit(x_train,y_train)
y_pred = text_model.predict(x_test)
y_pred
```

```python
import joblib 
joblib.dump(text_model,'sentiment12')
```
